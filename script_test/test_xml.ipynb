```python
import os, sys, importlib
import pandas as pd
pd.set_option('display.max_columns', None)
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
import xml.etree.ElementTree as ET
```

```python
def battery_xml_to_df(xml_file):
    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    data = []
    
    # Iterate through each timestep
    for timestep in root.findall('timestep'):
        t_time = timestep.get('time')
        
        # Iterate through each vehicle in that timestep
        for vehicle in timestep.findall('vehicle'):
            # Combine timestep info with vehicle attributes
            v_data = vehicle.attrib.copy()
            v_data['time'] = t_time
            data.append(v_data)
            
    # Create DataFrame and convert numeric columns
    df = pd.DataFrame(data)
    
    # Optional: Automatically convert strings to numbers where possible
    return df.apply(pd.to_numeric, errors='ignore')
```

```python
# Usage
model_dir = r"C:\Users\mmh\OneDrive - Oak Ridge National Laboratory\nashville_microsim\Model\SUMO\xml"
filename_battery = os.path.join(model_dir,"model baseline",'battery_stats.xml')
df = battery_xml_to_df(filename_battery)
```
