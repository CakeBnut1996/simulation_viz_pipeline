```python
import os, sys, importlib, duckdb
```

```python
db_path = os.path.join("..", "sumo_output.db")
con = duckdb.connect(db_path, read_only=True)
```

```python
metric = "meanSpeed"
job_id =  "20260127"
time_step = 8
```

```python
query = f"""
    SELECT 
        e.geometry_wkt, 
        e.ID as edge_id,
        r.id as detector_id,  -- Helpful to keep the original detector ID
        r.{metric} as value
    FROM all_e2 r
    JOIN edge e ON e.ID = regexp_replace(r.id, '_[^_]+$', '')
    WHERE r.sim_job_id = '{job_id}' 
    AND r.begin >= {time_step * 900} 
    AND r.begin < {(time_step + 1) * 900}
"""
df = con.execute(query).df()
df.head()
```
