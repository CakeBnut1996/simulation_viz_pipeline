```python
import os, sys, importlib, duckdb
```

```python
db_path = os.path.join("..", "sumo_output.db")
con = duckdb.connect(db_path, read_only=True)
```

```python
metric = "meanSpeed"
job_id =  "20260127"
time_step = 6
```

```python
query = f"""
    SELECT e.geometry_wkt, e.ID as edge_id, r.{metric} as value
    FROM edge e
    JOIN all_e2 r ON e.ID = r.id
    WHERE r.sim_job_id = '{job_id}' 
    AND r.begin >= {time_step * 900} 
    AND r.begin < {(time_step + 1) * 900}
"""
df = con.execute(query).df()
```
